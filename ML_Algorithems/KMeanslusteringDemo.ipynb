```python
import numpy as np
import pandas as pd
df = pd.read_csv('Datasets\student_clustering.csv')
print("The shape of data is",df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)
y_means = km.fit_predict(df)
```

```python
y_means
```

```python
df['cluster'] = y_means
```

```python
df[df['cluster'] == 1]['cgpa']
```

```python
plt.scatter(df[df['cluster'] == 0]['cgpa'],df[df['cluster'] == 0]['iq'])
plt.scatter(df[df['cluster'] == 1]['cgpa'],df[df['cluster'] == 1]['iq'])
plt.scatter(df[df['cluster'] == 2]['cgpa'],df[df['cluster'] == 2]['iq'])
plt.scatter(df[df['cluster'] == 3]['cgpa'],df[df['cluster'] == 3]['iq'])
```
